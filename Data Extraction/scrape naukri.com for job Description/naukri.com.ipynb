```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from urllib.request import urlopen 
import requests
import csv
import time
```

# Scrape Job Vacancy Links from naukri.com

```python
links=[]
```

```python
headers_std = {
'User-Agent': 'Mozilla/5.0 AppleWebKit/537.36 (KHTML, like Gecko; compatible; Googlebot/2.1; +http://www.google.com/bot.html) Safari/537.36',
}
for page in range(20,40):
    print("Scraping page=", page)
    url="https://www.naukri.com/engineering-jobs-"+str(page)
    html = requests.get(url,headers=headers_std).text
    soup = BeautifulSoup(html,'html.parser')
    table  = soup.findAll("a",{ "class" : "ellipsis" })
    for i in table:
        l=i.get('href')
        if l.startswith("https://www.naukri.com/job-listings-"):
            links.append(l)
            print(l)
            
    
##Google has blocked my bot, tried changing username but no avail
```

# Create a new file for storing details

```python
f=open('job details.csv', 'a', newline='')
fieldnames = ['Job Title','Company Name', 'Job Location', 'Job Description', 'Job Link']
writer = csv.DictWriter(f, fieldnames=fieldnames)
#writer.writeheader()
```

# Func to check if object is not 'None'

```python
def validate_field(field):
    if field==None:
       field = 'No results'
    try:
        field=field.text
    except:
        print("Error")
    return field
```

# Copy Links to links list

```python
with open('Job links.csv') as f:            ##profile links.csv in the same directory as that of program
    reader = csv.reader(f)
    links = np.array(list(reader)).flatten()
```

# Extract job details from each link

```python
headers_std = {
'User-Agent': 'Mozilla/5.0 AppleWebKit/537.36 (KHTML, like Gecko; compatible; Googlebot/2.1; +http://www.google.com/bot.html) Safari/537.36',
}
#.dang-inner-html div
start=50
n=start
for l in links[start:]:
    print("Fetching information file-", n)
    n=n+1
    url=l
    html = requests.get(url,headers=headers_std).text
    soup = BeautifulSoup(html,'html.parser')
    body = soup.body
    for br in body.find_all("br"):
        br.replace_with("\n")
    job_title=validate_field(body.find("h1",{ "class" : "jd-header-title" }))
    job_location=validate_field(body.find("span",{ "class" : "location" }))
    job_link=url
    job_description=""""""""
    company_name=validate_field(body.find("div",{ "class" : "jd-header-comp-name" }))
    for paragraph in body.findAll("div",{ "class" : "dang-inner-html" }):
        job_description=job_description+"\n"+paragraph.text
    writer.writerow({'Job Title': job_title, 'Company Name': company_name, 'Job Location': job_location,  'Job Description': job_description, 'Job Link':job_link})
f.close()
```
